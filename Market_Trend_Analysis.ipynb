```python
def group_industries(industry_name):
    """
    Groups industries into broader categories based on a predefined mapping.
    
    Args:
        industry_name (str): The name of the industry to be grouped.
    
    Returns:
        str: The broader category the industry belongs to.
    """
    
    # Return the broader category for the given industry name
    return industry_mapping.get(industry_name, 'Other')  # Default to 'Other' if not found
```

```python
# Find missing industries
missing_industries = [industry for industry in unique_industries if industry not in industry_mapping]
print("Missing industries:", missing_industries)
```

```python
# Apply the grouping to a new column
data_cleaned['Industry Category'] = data_cleaned['Industry Name'].apply(group_industries)
data_cleaned[['Industry Name', 'Industry Category']]
```

```python
starting_industry = data_cleaned.groupby(['Industry Category','Employment Start Year']).size().reset_index(name='Count')

starting_industry
```

```python
# Loop through each industry and plot its trend
for industry in starting_industry['Industry Category'].unique():
    # Getting the data for the current industry
    industry_data = starting_industry[starting_industry['Industry Category'] == industry]
    plt.plot(
        industry_data['Employment Start Year'],
        industry_data['Count'],
        label=industry
    )

plt.xlabel('Employment Start Year')
plt.ylabel('Count')
plt.title('Employment Trends by Industry')
plt.legend(title='Industry Name')
plt.grid(True)
plt.show()
```

```python
total_counts = starting_industry.groupby('Industry Category')['Count'].sum().sort_values(ascending=False)
print(total_counts)
# Get the top 6 and bottom 6 industries
top_categories = total_counts.head(6).index.tolist()
bottom_categories = total_counts.tail(6).index.tolist()
# Filter for top and bottom categories
top_data = starting_industry[starting_industry['Industry Category'].isin(top_categories)]
bottom_data = starting_industry[starting_industry['Industry Category'].isin(bottom_categories)]
```

```python
# Plot for top categories
plt.figure(figsize=(15, 13))
for category in top_categories:
    subset = top_data[top_data['Industry Category'] == category]
    plt.plot(subset['Employment Start Year'], subset['Count'], label=category)

plt.title('Employment Trends by Industry (Top 6 Categories)')
plt.xlabel('Employment Start Year')
plt.ylabel('Count')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Place legend outside
plt.tight_layout()
plt.show()
```

```python
# Plot for bottom categories
plt.figure(figsize=(15, 13))
for category in bottom_categories:
    subset = bottom_data[bottom_data['Industry Category'] == category]
    plt.plot(subset['Employment Start Year'], subset['Count'], label=category)

plt.title('Employment Trends by Industry (Bottom 6 Categories)')
plt.xlabel('Employment Start Year')
plt.ylabel('Count')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Place legend outside
plt.tight_layout()
plt.show()
```

```python
# Data Forecasting
from prophet import Prophet

# Convert 'Employment Start Year' to datetime
starting_industry['Employment Start Year'] = pd.to_datetime(starting_industry['Employment Start Year'], format='%Y')

unique_industries = starting_industry['Industry Category'].unique()

for industry in unique_industries:
    # Filter data for the current industry
    industry_category = starting_industry[starting_industry['Industry Category'] == industry]
    
    # Prepare data for Prophet
    prophet_data = industry_category[['Employment Start Year', 'Count']]
    prophet_data.columns = ['ds', 'y']
    
    # Fit the model
    model = Prophet()
    model.fit(prophet_data)
    
    # Create a future dataframe
    future = model.make_future_dataframe(periods=5, freq='Y')
    forecast = model.predict(future)
    
    # Plot the forecast
    model.plot(forecast)
    plt.title(f'Forecast for {industry} Industry Growth', fontsize=16)
    plt.xlabel('Year', fontsize=14)
    plt.ylabel('Count', fontsize=14)
```
